```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Prompt the user to select a file.
uploaded = files.upload()

# Read the uploaded file into a DataFrame.
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Print the uploaded file's name.
print(file_name)

# Print the DataFrame.
print(df.head())

# Computing X and Y
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values

# Calculate Parameters
X_sum = np.sum(X)
X_squared_sum = np.sum(X * X)
n = len(X)
print("Sum of X:", X_sum)
print("Sum of X squared:", X_squared_sum)
print("Length of X:", n)

# Estimate Parameters
numer = n * np.sum(X * y) - np.sum(X) * np.sum(y)
denom = n * np.sum(X * X) - (np.sum(X)) ** 2
w1 = numer / denom
w0 = (np.sum(y) - w1 * (np.sum(X))) / n
print("w0:", w0)
print("w1:", w1)

# Plot Regression Line
max_x = np.max(X)
min_x = np.min(X)
x1 = np.linspace(min_x, max_x)
y1 = w0 + w1 * x1

plt.plot(x1, y1, color='red', label='Regression Line')
plt.scatter(X, y, c='green', label='Scatter Plot')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()

# Calculate RMSE
y_pred = w0 + w1 * X
rmse = np.sqrt(np.mean((y - y_pred) ** 2))
print("RMSE:", rmse)

# Calculate R2 Score
ss_tot = np.sum((y - np.mean(y)) ** 2)
ss_res = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_res / ss_tot)
print("R2 Score:", r2)
```

## **Uni-variate Regression**

```python
import matplotlib.pyplot as plt
from scipy import stats
# Input data
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
# Perform Uni varient regression
slope, intercept, r, p, std_err = stats.linregress(x, y)
# Define a function for the regression line
def myfunc(x):
  return slope * x + intercept
# Generate predicted values using the regression line
mymodel = list(map(myfunc, x))
#Uni-variate Regression
# Plot the data points and the regression line
plt.scatter(x, y)
plt.plot(x, mymodel)
# Print R-squared value
print("R Squared Value:", r)
# Print standard error value
print("std_err Value:", std_err)
# Show the plot
plt.show()
```

```python
from sklearn.neighbors import RadiusNeighborsClassifier

# Data
X = [[1, 0], [1.2, 0.2], [2.5, 1], [3, 1.5], [1, 0.5], [2.5, 2], [2.3, 2.7]]
y = [0, 0, 1, 1, 0, 1, 1]

# Try different values of radius
radius_values = [0.5, 1.0, 1.5]

for radius in radius_values:
    # Define and fit the classifier
    classifier = RadiusNeighborsClassifier(radius=radius)
    classifier.fit(X, y)

    # Test points
    test_points = [[1.1, 0.1], [2.7, 2.5]]

    # Predict the class labels for the provided data
    predictions = classifier.predict(test_points)

    print(f"For radius={radius}:")
    for i, point in enumerate(test_points):
        print(f"Point {i+1}: {point} -> Predicted Class: {predictions[i]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv('headbrain.csv')
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values

# Calculate w0 and w1 using formula
X_sum = np.sum(X)
y_sum = np.sum(y)
X_square_sum = np.sum(X**2)
Xy_sum = np.sum(X*y)
n = len(X)

slope = (n*Xy_sum - X_sum*y_sum) / (n*X_square_sum - X_sum**2)
intercept = (y_sum - slope*X_sum) / n

# Plot data and regression line
plt.scatter(X, y)
plt.plot(X, slope*X + intercept, color='red')
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
plt.title('Regression Line')
plt.show()

# Calculate RMSE
rmse = np.sqrt(np.mean((y - (slope*X + intercept))**2))
print("RMSE:", rmse)

# Calculate R2
y_mean = np.mean(y)
ss_tot = np.sum((y - y_mean)**2)
ss_res = np.sum((y - (slope*X + intercept))**2)
r2 = 1 - (ss_res / ss_tot)
print("R2:", r2)

# Plot the data points
plt.scatter(X, y)
plt.xlabel('Head Size (cm^3)')
plt.ylabel('Brain Weight (grams)')
plt.title('Data Points')
plt.show()
```
