```python
# Import standard Libraries
import pandas as pd
import sklearn
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
```

```python
# Load wine dataset
wine = load_wine()
```

```python
# Dataset description
print("Description:", wine.keys())

# Names of classes
print("\nTarget Names:", wine.target_names)

# Features
print(f"\nFeature Name: {wine.feature_names}")

# Shape
print("\nShape / Dimension of Input:", wine.data.shape)
print("\nShape / Dimension of Output:", wine.target.shape)
n_samples, n_features =  wine.data.shape
print(f"\nNumber of samples: {n_samples}")
print(f"\nNumber of features: {n_features}")

# Display first 5 rows of Input and Output
print(f"\nFirst 5 rows of Input: \n{wine.data[:5]}")
print(f"\nFirst 5 values of Output: {wine.target[:5]}")
```

```python
# Create Histogram of dataset
fig, ax = plt.subplots()
x_index = 3
colors = ["blue", "red", "green"]

for label, color in zip(range(len(wine.target_names)), colors):
  ax.hist(wine.data[wine.target==label, x_index],
          label=wine.target_names[label],
          color=color)

  ax.set_xlabel(wine.feature_names[x_index])
  ax.legend(loc="upper right")

fig.show()
```

```python
# Scatter Plot
fig, ax = plt.subplots()
x_index = 3
y_index = 0
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(wine.target_names)), colors):
  ax.scatter(wine.data[wine.target == label, x_index], wine.data[wine.target == label, y_index], label=wine.target_names[label], c=color)
  ax.set_xlabel(wine.feature_names[x_index])
  ax.set_ylabel(wine.feature_names[y_index])
  ax.legend(loc='upper right')
plt.show()
```

```python
# Create differnt subplots
n = len(wine.feature_names)
fig, ax = plt.subplots(n, n, figsize=(50, 50))
colors = ['blue', 'red', 'green']

for x in range(n):
  for y in range(n):
    xname = wine.feature_names[x]
    yname = wine.feature_names[y]

    for color_ind in range(len(wine.target_names)):
        ax[x, y].scatter(wine.data[wine.target==color_ind,x],wine.data[wine.target==color_ind, y],label=wine.target_names[color_ind],c=colors[color_ind])
        ax[x, y].set_xlabel(xname)
        ax[x, y].set_ylabel(yname)
        ax[x, y].legend(loc='upper left')
plt.show()
```

```python
# Convert into Pandas DataFrame
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['Target']=wine.target

# Print first 5 rows of DataFrame
print(wine_df.head(5))

# Display Info
print(wine_df.info())

# Convert target datatype from int to category-object
wine_df['Target'] = wine_df['Target'].astype(object)
print(wine_df.info())

print(wine_df.groupby('Target').size())
```
